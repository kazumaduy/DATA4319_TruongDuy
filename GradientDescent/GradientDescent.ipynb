# Gradient Descent 

Optimization and minimization techniques are often used in machine learning to help determine the best cost of a function. Gradient descent is an iterative method that continuously updates our minimizers over our specified function.

```julia
using Plots
theme(:solarized)


# Define a simple function to explore minimization on.
C(w) = (w - 4)^3 - 0.5w^2


plot(C,
    color = "cyan",
    xlims = (0,8),
    xaxis = "w",
    ylims = (-15, 5),
    yaxis = "Cost C(w)",
    legend = false)
```

#### Global minimization takes time and therefore we must make an educated guess as to where our initial point may be. For example we shall take a random integer as our guess and see how it follows.

```julia
w_0 = rand((4:7))
println("---------")
println("Our initial guess is $w_0")
println("---------")
scatter!([(w_0, C(w_0))], 
        color = "white")
```

#### From the plot above we must decide a direction we want to go in order to obtain the global minimization, which to us mathematically and visually, lies between 4 and 7.

```julia
# Define the derivative of C(w)
∇C(w) = (3w-12)^2 - w

# Set your step length
α = .1

# Apply gradient descent 
w_1 = w_0 - α*∇C(w_0)


println("w_0 = $w_0 and |∇C(w_0)| = ", abs(∇C(w_0)))
println("")
println("w_1 = $w_1 and |∇C(w_1)| = ", abs(∇C(w_1)))
println("")

plot!([(w_0, C(w_0)), (w_1, C(w_1))],
        color = "red")
scatter!([(w_0, C(w_0)), (w_1, C(w_1))],
        color = "white")
```

```julia
w_2 = w_1 - α*∇C(w_1)

println("w_0 = $w_0 and |∇C(w_0)| = ", abs(∇C(w_0)))
println("")
println("w_1 = $w_1 and |∇C(w_1)| = ", abs(∇C(w_1)))
println("")
println("w_2 = $w_2 and |∇C(w_2)| = ", abs(∇C(w_2)))
println("")

plot!([(w_0, C(w_0)), (w_1, C(w_1)), (w_2, C(w_2))],
        color = "red")
scatter!([(w_0, C(w_0)), (w_1, C(w_1)), (w_2, C(w_2))],
        color = "white")
```

```julia
w_3 = w_2 - α*∇C(w_2)

println("w_0 = $w_0 and |∇C(w_0)| = ", abs(∇C(w_0)))
println("")
println("w_1 = $w_1 and |∇C(w_1)| = ", abs(∇C(w_1)))
println("")
println("w_2 = $w_2 and |∇C(w_2)| = ", abs(∇C(w_2)))
println("")
println("w_3 = $w_3 and |∇C(w_3)| = ", abs(∇C(w_3)))
println("")

plot!([(w_0, C(w_0)), (w_1, C(w_1)), (w_2, C(w_2)), (w_3, C(w_3))],
        color = "red")
scatter!([(w_0, C(w_0)), (w_1, C(w_1)), (w_2, C(w_2)), (w_3, C(w_3))],
        color = "white")
```

```julia
w_4 = w_3 - α*∇C(w_3)

println("w_0 = $w_0 and |∇C(w_0)| = ", abs(∇C(w_0)))
println("")
println("w_1 = $w_1 and |∇C(w_1)| = ", abs(∇C(w_1)))
println("")
println("w_2 = $w_2 and |∇C(w_2)| = ", abs(∇C(w_2)))
println("")
println("w_3 = $w_3 and |∇C(w_3)| = ", abs(∇C(w_3)))
println("")
println("w_4 = $w_4 and |∇C(w_4)| = ", abs(∇C(w_4)))
println("")

plot!([(w_0, C(w_0)), (w_1, C(w_1)), (w_2, C(w_2)), (w_3, C(w_3)), (w_4, C(w_4))],
        color = "lightblue")
scatter!([(w_0, C(w_0)), (w_1, C(w_1)), (w_2, C(w_2)), (w_3, C(w_3)), (w_4, C(w_4))],
        color = "white")
```

#### As we iterate the function further and further our gradient descent reaches closer to 0 meaning we are reaching a local minima
